**Encuentre el polinomio de Lagrange para los siguientes datos y grafique:**

(0,0),(30,0.5),(60,3/√2),(90,1)

(1,1),(2,2),(3,2)

(−2,5),(1,7),(3,11),(7,34)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

# Definir los conjuntos de datos
datasets = [
    [(0, 0), (30, 0.5), (60, np.sqrt(3)/2), (90, 1)],
    [(1, 1), (2, 2), (3, 2)],
    [(-2, 5), (1, 7), (3, 11), (7, 34)]
]

# Crear una figura para graficar
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Generar y graficar los polinomios de Lagrange
polynomials = []

for i, data in enumerate(datasets):
    x_vals = np.array([point[0] for point in data])
    y_vals = np.array([point[1] for point in data])
    
    # Calcular el polinomio de Lagrange
    poly = lagrange(x_vals, y_vals)
    polynomials.append(poly)
    
    # Crear puntos para la gráfica
    x_plot = np.linspace(min(x_vals) - 5, max(x_vals) + 5, 400)
    y_plot = poly(x_plot)
    
    # Graficar
    axs[i].plot(x_plot, y_plot, label=f'Polinomio {i+1}', color='orange')
    axs[i].scatter(x_vals, y_vals, color='red')
    axs[i].set_title(f'Polinomio de Lagrange {i+1}')
    axs[i].grid(True)
    axs[i].legend()

plt.tight_layout()
plt.show()

# Mostrar expresiones de los polinomios
for i, poly in enumerate(polynomials, start=1):
    print(f"Polinomio {i}:\n{poly}\n")
```
